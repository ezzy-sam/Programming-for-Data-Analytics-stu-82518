```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Sales Data_PDA_4052.xlsx'
df = pd.read_excel(file_path,header=1)

df.rename(columns={'value_': 'value'}, inplace=True)

# Correct Data Types: Convert 'date' to datetime for time-series analysis.
df['date'] = pd.to_datetime(df['date'])

print("--- 1. INITIAL INSPECTION (Task 1.1) ---")
print(f"Data Loaded: {df.shape[0]} initial records.")
print("Data Types after correction:")
print(df.info())

# PART 2: PRE-PROCESSING & AGGREGATION (Task 1.2)

order_agg = df.groupby('order_id').agg(
    total_order_value=('value', 'sum'),
    sales_person=('sales_person', lambda x: x.mode()[0]),
    priority=('priority', lambda x: x.mode()[0]),         
    date=('date', 'first')                               
).reset_index()

print("\n--- 2. PRE-PROCESSING COMPLETE (Task 1.2) ---")
print(f"Aggregated to {order_agg.shape[0]} unique orders.")
print("New Descriptive Statistics (Total Order Value):")
print(order_agg['total_order_value'].describe())

# PART 3: INVESTIGATION A: SALES PERSON PERFORMANCE 

person_sales = order_agg.groupby('sales_person')['total_order_value'].sum().sort_values(ascending=False)

# Plot simple bar chart
plt.figure(figsize=(7, 5))
person_sales.plot(kind='bar', color=['#3B82F6', '#10B981', '#F59E0B'])
plt.title('Total Sales Value by Sales Person')
plt.ylabel('Total Value ()')
plt.xlabel('Sales Person')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.5)
plt.tight_layout()
plt.savefig('graph_1_sales_person_total.png') 
plt.show()

print("\n3. SALES PERSON TOTALS")
print(person_sales)

# PART 4: INVESTIGATION B: PRIORITY VS. VALUE CORRELATION 

# Box Plot visualises distribution, median, and outliers for correlation.
plt.figure(figsize=(8, 6))
order_agg.boxplot(column='total_order_value', by='priority', grid=True)
plt.title('Order Value Distribution by Priority')
plt.suptitle('') # Remove automatic subtitle
plt.ylabel('Total Value ()')
plt.xlabel('Order Priority')
plt.tight_layout()
plt.savefig('graph_2_priority_correlation.png')
plt.show()

print("\n 4. AVERAGE VALUE BY PRIORITY")
print(order_agg.groupby('priority')['total_order_value'].mean().sort_values(ascending=False))

# PART 5: INVESTIGATION C: TIME TRENDS
# Create a month column for grouping
order_agg['Month'] = order_agg['date'].dt.to_period('M')

# Group by Month and Sales Person
monthly_trends = order_agg.groupby(['Month', 'sales_person'])['total_order_value'].sum().unstack()

# Generate Visualisation (Line Chart)
plt.figure(figsize=(10, 6))
monthly_trends.plot(kind='line', marker='o', linewidth=2, ax=plt.gca())
plt.title('Monthly Sales Trends by Sales Person')
plt.ylabel('Total Value ()')
plt.xlabel('Month')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Sales Person')
plt.tight_layout()
plt.savefig('graph_3_monthly_trends.png')
plt.show()

print("\n5. MONTHLY TRENDS")
print(monthly_trends)
```
